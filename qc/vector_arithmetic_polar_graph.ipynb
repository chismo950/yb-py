```python
# Vector arithmetic and graphical display (no input version)

from math import atan2, degrees
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display

# Step 1: Fixed vectors
v1 = np.array([3, 4], dtype=float)
v2 = np.array([1, 2], dtype=float)

# Step 2: Convert to polar coordinates
def to_polar(vec):
    r = np.linalg.norm(vec)
    theta = degrees(atan2(vec[1], vec[0]))
    return r, theta

r1, theta1 = to_polar(v1)
r2, theta2 = to_polar(v2)

print(f"Vector A in polar coordinates: (r={r1:.2f}, θ={theta1:.2f}°)")
print(f"Vector B in polar coordinates: (r={r2:.2f}, θ={theta2:.2f}°)")

# Step 3: Various operations
add = v1 + v2
sub = v1 - v2
div = np.divide(v1, v2, out=np.zeros_like(v1, dtype=float), where=v2!=0)
mul = v1 * v2
power = np.power(v1, v2)

# Step 4: Plotting function
def plot_vectors(vecs, labels, title):
    plt.figure(figsize=(6, 6))
    origin = np.zeros(2)
    colors = ['r', 'g', 'b', 'c', 'm', 'y']
    for i, v in enumerate(vecs):
        plt.quiver(*origin, *v, angles='xy', scale_units='xy', scale=1, color=colors[i%len(colors)], label=labels[i])
        plt.text(v[0]*1.05, v[1]*1.05, labels[i], fontsize=12)
    plt.xlim(-5, 10)
    plt.ylim(-5, 10)
    plt.grid()
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.legend()
    plt.title(title)
    fig = plt.gcf()
    display(fig)
    plt.close(fig)

# Original vectors
plot_vectors([v1, v2], ["A", "B"], "Original vectors A and B")
# Operation result plots
plot_vectors([add], ["A+B"], "Addition (A+B)")
plot_vectors([sub], ["A-B"], "Subtraction (A-B)")
plot_vectors([div], ["A/B"], "Element-wise Division (A/B)")
plot_vectors([mul], ["A*B"], "Element-wise Multiplication (A*B)")
plot_vectors([power], ["A^B"], "Element-wise Power (A^B)")

# Output all results
print("==== Resultant vectors (components) ====")
print(f"A+B: {add}")
print(f"A-B: {sub}")
print(f"A/B: {div}")
print(f"A*B: {mul}")
print(f"A^B: {power}")

print("\n==== Polar coordinates ====")
print(f"A (r={r1:.2f}, θ={theta1:.2f}°)")
print(f"B (r={r2:.2f}, θ={theta2:.2f}°)")
```
